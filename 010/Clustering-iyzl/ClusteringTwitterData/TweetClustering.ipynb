```python
import pandas as pd
import glob
import os
```

```python
path = r'Tweets'
all_files = glob.glob(os.path.join(path , "*.txt"))
all_files
```

```python
li = []
for filename in all_files:
    try:
        df = pd.read_csv(filename, sep='|', header=None, names=['ID', 'Date', 'Tweet'])
        li.append(df)
    except:
        continue
data = pd.concat(li, axis=0, ignore_index=True)
data.head()
```

```python
data.info()
```

```python
X = data[['Tweet']]
X.head()
```

```python
with open('stopwords.txt', encoding='utf8') as stopwords_file:
    stopwords = stopwords_file.readlines()
stopwords = [line.replace('\n', '') for line in stopwords]
stopwords
```

```python
from nltk import word_tokenize
import nltk
import re
nltk.download('punkt')
```

```python
dataset = pd.DataFrame(columns=['title_body'])
for index, row in X.iterrows():
    title_body_tokenized = word_tokenize(row['Tweet'])
    title_body_tokenized_filtered = [w.lower() for w in title_body_tokenized if not w.lower() in stopwords]
    text = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', " ", ' '.join(title_body_tokenized_filtered))
    text = re.sub('[^\w\s]', '', text)
    text = re.sub("\d+", "", text)
    dataset.loc[index] = {
        'title_body': text,
    }
```

```python
dataset
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(dataset['title_body'])
```

```python
X = vectorizer.transform(dataset['title_body'])
```

```python
X
```

```python
import umap
reducer = umap.UMAP(n_components=1000)
```

```python
X_reduced = reducer.fit_transform(X)
```

```python
import numpy as np
np.shape(X_reduced)
```

```python
from sklearn.cluster import MeanShift, SpectralClustering, KMeans
```

```python
msh = KMeans(n_clusters=6)
msh.fit(X_reduced)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, msh.labels_)
```

```python
dataset['Cluster'] = msh.labels_
```
